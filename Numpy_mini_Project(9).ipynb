```python
import numpy as np
prices=np.array([100,200,250,50,300])
quantity=np.array([1,2,2,4,1])
discount=np.array([10,20,0,5,50])
total=np.multiply(prices,quantity)
print("Item totals:",total)
final=np.subtract(total,discount)
print("After discount:",final)
bill=np.add.reduce(final)
print("Final bill:",bill)
```

```python
import numpy as np
pins=np.array([1,4,7,2,9,5])
def score(x):
    return x*3+7
ufunc_score=np.frompyfunc(score,1,1)
result=ufunc_score(pins)
print("Scores:",result)
print("Max:",np.max(result))
print("Min:",np.min(result))
```
