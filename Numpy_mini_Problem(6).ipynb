```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
lambda_rate=8
hours=500
arrivals=np.random.poisson(lam=lambda_rate,size=hours)
print("Average customers per hour:",np.mean(arrivals))
print("Max customers in an hour:",np.max(arrivals))
print("Min customers in an hour:",np.min(arrivals))
less5=np.sum(arrivals<5)
between5_10=np.sum((arrivals>=5)&(arrivals<=10))
more10=np.sum(arrivals>10)
print("\nHours with <5 customers:",less5)
print("Hours with 5-10 customers:",between5_10)
print("Hours with >10 customers:",more10)
sns.histplot(arrivals,kde=True)
plt.title("Customer Arrival Distribution")
plt.xlabel("Number of Customers per Hour")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
num_trips=200
min_time=1
max_time=5
travel_times=np.random.uniform(low=min_time,high=max_time,size=num_trips)
print("Average travel time:",np.mean(travel_times))
print("Minimum travel time:",np.min(travel_times))
print("Maximum travel time:",np.max(travel_times))
less2=np.sum(travel_times<2)
between2_4=np.sum((travel_times>=2)&(travel_times<=4))
more4=np.sum(travel_times>4)
print("\nTrips<2 hours:",less2)
print("Trips 2-4 hours:",between2_4)
print("Trips>4 hours:",more4)
sns.histplot(travel_times,kde=True)
plt.title("Travel Time Distribution")
plt.xlabel("Travel Time")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
num_customers=500
mean_score=5
scale=1.5
scores=np.random.logistic(loc=mean_score,scale=scale,size=num_customers)
scores=np.round(scores,1)
print("Average score:",np.mean(scores))
print("Maximum score:",np.max(scores))
print("Minimum score:",np.min(scores))
low=np.sum(scores<4)
medium=np.sum((scores>=4)&(scores<=7))
high=np.sum(scores>7)
print("\nLow satisfaction:",low)
print("Medium satisfaction:",medium)
print("High satisfaction:",high)
sns.histplot(scores,kde=True)
plt.title("Cutomer Satisfaction Distribution")
plt.xlabel("Satisfaction Score")
plt.ylabel("Number of Customers")
plt.show()
```
