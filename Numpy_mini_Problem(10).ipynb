```python
import numpy as np
prices=np.array([99.49,150.75,199.99,49.25,20.50])
print("Original:",prices)
around=np.around(prices)
floor=np.floor(prices)
ceil=np.ceil(prices)
trunc=np.trunc(prices)
print("\nAround:",around)
print("Floor:",floor)
print("Ceil:",ceil)
print("Trunc:",trunc)
print("\nTotal(around):",np.sum(around))
print("Total(floor):",np.sum(floor))
print("Total(ceil):",np.sum(ceil))
print("Total(trunc):",np.sum(trunc))
```

```python
import numpy as np
intensity=np.array([10,100,1000,10000])
log_n=np.log(intensity)
log_2=np.log2(intensity)
log_10=np.log10(intensity)
print("Natural Log:",log_n)
print("Log2:",log_2)
print("Log10:",log_10)
print("Max log10:",np.max(log_10))
```

```python
import numpy as np
steps=np.array([3500,4200,5000,6200,3000,4000,4500])
total1=np.add.reduce(steps)
total2=np.sum(steps)
cum=np.cumsum(steps)
print("Total(reduce):",total1)
print("Total(sum):",total2)
print("\nCumulative:",cum)
day=np.argmax(cum>15000)+1
print("\nReached 15000 on day:",day)
```
