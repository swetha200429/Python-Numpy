```python
import numpy as np
growth=np.array([1.05,1.10,0.95,1.20,1.08])
p1=np.prod(growth)
p2=np.multiply.reduce(growth)
cum=np.cumprod(growth)
print("Total growth (prod):",p1)
print("Total growth (reduce):",p2)
print("\nCumulative growth:")
print(cum)
start=1000
final=start*p1
print("\nFinal money:",round(final,2))
```

```python
import numpy as np
temps=np.array([30,32,31,35,40,38,39])
change=np.diff(temps)
print("Changes:",change)
print("Max increase:",np.max(change))
print("Max decrease:",np.min(change))
increase_days=np.sum(change>0)
print("Increase days:",increase_days)
```

```python
import numpy as np
signals=np.array([30,45,60,90])
l1=np.lcm(signals[0],signals[1])
print("LCM of first two:",l1)
l_all=np.lcm.reduce(signals)
print("\nLCM of all signals:",l_all)
print("\nAll signals match after",l_all,"second")
```
