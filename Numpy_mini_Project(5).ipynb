```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
speed=np.random.normal(loc=50,scale=10,size=1000)
print("Average:",np.mean(speed))
print("Max:",np.max(speed))
print("Min:",np.min(speed))
sns.histplot(speed,kde=True)
plt.title("Internet Speed Distribution")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sleep=np.random.normal(loc=7,scale=1.5,size=500)
print("Average:",np.mean(sleep))
print("Max:",np.max(sleep))
print("Min:",np.min(sleep))
less5=np.sum(sleep<5)
between=np.sum((sleep>=5)&(sleep<=9))
more9=np.sum(sleep>9)
print("\nLess than 5 hours:",less5)
print("5 to 9 hours:",between)
print("More than 9 hours:",more9)
sns.histplot(sleep,kde=True)
plt.title("Sleep Hours Normal Distribution")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=np.random.binomial(n=50,p=0.85,size=1000)
print("Average:",np.mean(data))
print("Max:",np.max(data))
print("Min:",np.min(data))
less40=np.sum(data<40)
mid=np.sum((data>=40)&(data<=45))
more45=np.sum(data>45)
print("\nLess than 40:",less40)
print("40 to 45:",mid)
print("More than 45:",more45)
sns.histplot(data, kde=True)
plt.title("Delivery Success Distribution")
plt.show()
```
