```python
import numpy as np
temp=np.random.randint(20,41,size=30)
print("Temperatures:\n",temp)
print("\nMax:",np.max(temp))
print("Min:",np.mean(temp))
print("Average:",np.mean(temp))
hot=np.sum(temp>=35)
normal=np.sum((temp>=25)&(temp<35))
cold=np.sum(temp<25)
print("\nHot days:",hot)
print("Normal days:",normal)
print("Cold days:",cold)
```

```python
import numpy as np
visitors = np.random.normal(loc=500, scale=50, size=7).astype(int)
clicks = np.random.binomial(n=100, p=0.3, size=7)
errors = np.random.poisson(lam=5, size=7)
print("Visitors:", visitors)
print("Ad Clicks:", clicks)
print("Errors:", errors)
print("\nHighest visitors:", np.max(visitors))
print("Total clicks:", np.sum(clicks))
print("Total errors:", np.sum(errors))
```

```python
import numpy as np
playlist=np.array([
    "Song1","Song2","Song3","Song4","Song5","Song6","Song7","Song8"
])
print("Original playlist:")
print(playlist)
perm=np.random.permutation(playlist)
print(perm)
np.random.shuffle(playlist)
print("\nShuffled:")
print(playlist)
print("\nIs changed:",not np.array_equal(playlist,perm))
```
