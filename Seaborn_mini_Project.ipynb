```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("shopping_data.csv")
print(df.head())
print(df.info())
sns.scatterplot(x="Spending_Score",y="Purchase_Amount",data=df)
plt.title("Spending Score vs Purchase Amount")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
calories=[200,250,230,300,350,400,380]
sns.set_style("darkgrid")
sns.set_context("talk")
sns.lineplot(x=days,y=calories,marker='o',c='red',mec='blue',mfc='blue')
plt.title("Weekly Calories Burned")
plt.show()
```
