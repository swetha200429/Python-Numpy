```python
import numpy as np
intervals=np.array([24,36,60,48])
g1=np.gcd(intervals[0],intervals[1])
print("GCD of first two:",g1)
g_all=np.gcd.reduce(intervals)
print("\nGCD of all intervals:",g_all)
print("\nCommon time unit:",g_all,"minutes")
```

```python
import numpy as np
angles=np.array([0,30,45,60,90])
distance=10
rad=np.deg2rad(angles)
x=distance*np.cos(rad)
y=distance*np.sin(rad)
print("X movement:",x)
print("Y movement:",y)
print("\nTotal X:",np.sum(x))
print("Total Y:",np.sum(y))
```

```python
import numpy as np
x=np.array([-2,-1,1,2])
height=np.cosh(x)
slope=np.sinh(x)
norm=np.tanh(x)
print("Height:",height)
print("Slope:",slope)
print("Normlized slope:",norm)
print("\nMax height:",np.max(height))
```

```python
import numpy as np
users_A=np.array([101,102,103,104,105,106])
users_B=np.array([104,105,106,107,108])
A=np.unique(users_A)
B=np.unique(users_B)
print("Unique A:",A)
print("Unique B:",B)
print("\nAll users:",np.union1d(A,B))
print("\nCommon users:",np.intersect1d(A,B))
print("\nOnly A:",np.setdiff1d(A,B))
print("\nOnly one platform:",np.setxor1d(A,B))
```
