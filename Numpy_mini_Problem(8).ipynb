```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
devices=500
scale=2.0
signal=np.random.rayleigh(scale=scale,size=devices)
print("Average signal:",np.mean(signal))
print("Max signal:",np.max(signal))
print("Min signal:",np.min(signal))
weak=np.sum(signal<1.5)
medium=np.sum((signal>=1.5)&(signal<=3))
strong=np.sum(signal>3)
print("\nWeak signal:",weak)
print("Medium signal:",medium)
print("Strong signal:",strong)
sns.histplot(signal,kde=True)
plt.title("Signal Strength Distribution")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
files=500
shape=3
sizes=np.random.pareto(a=shape,size=files)+1
print("Average size:", np.mean(sizes))
print("Max size:",np.max(sizes))
print("Min size:",np.min(sizes))
small=np.sum(sizes<2)
medium=np.sum((sizes>=2)&(sizes<=4))
large=np.sum(sizes>4)
print("\nSmall files:",small)
print("Medium files:",medium)
print("Large files:",large)
sns.histplot(sizes,kde=True)
plt.title("File Size Distribution")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
videos=500
a=5
ranks=np.random.zipf(a=a,size=videos)
print("Average rank:",np.mean(ranks))
print("Max rank:",np.max(ranks))
print("Min rank:",np.min(ranks))
popular=np.sum((ranks==1)|(ranks==2))
medium=np.sum((ranks>=3)&(ranks<=5))
low=np.sum(ranks>5)
print("\nPopular:",popular)
print("Medium:",medium)
print("Low:",low)
sns.histplot(ranks,kde=False)
plt.title("Video Popularity Distribution")
plt.show()
```
