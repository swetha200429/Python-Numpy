```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
study_hours=[2,3,4,3.5,5,6,4]
plt.plot(days,study_hours,c='hotpink')
plt.title("Weekly Study Hours",fontsize=14,color='blue')
plt.xlabel("Days of the Week",fontsize=12,color='green')
plt.ylabel("Hours Studied",fontsize=12,color='red')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
temperature=[30,32,31,33,35,36,34]
plt.plot(days,temperature,marker='s',c='green')
plt.title("Weekly Temperature Variation")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.grid(True,ls='--',lw=0.5)
max_temp=max(temperature)
max_day = days[temperature.index(max_temp)]
plt.text(max_day,max_temp,f"Peak:{max_temp}°C")
plt.show()
```
