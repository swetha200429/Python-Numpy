```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
study_hours=[2,3,4,3.5,5,6,4]
exercise_hours=[1,1.5,1,2,2,2.5,1.5]
plt.subplot(2,1,1)
plt.plot(days,study_hours, color='darkblue',marker='D',mec='r')
plt.title("Daily Study Hours")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.subplot(2,1,2)
plt.plot(days,exercise_hours,color='darkblue',marker='D',mec='r')
plt.title("Daily Exercise Hours")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
screen_time=[1,2,3,4,5,6,7]
sleep_hours=[8,7.5,7,6.5,6,5.5,5]
plt.scatter(screen_time,sleep_hours,color='blue',s=80)
plt.title("Screen Time vs Sleep Hours")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Sleep Hours")
plt.grid()
plt.show()
```
